## 合作次數

```python
import pandas as pd

idol_name = pd.read_excel("D:/大學/四下/社群媒體分析/報告/K-POP藝人清單.xlsx", sheet_name = "idols")
k_name = idol_name["name (korean)"].tolist()
e_name = idol_name["name (english)"].tolist()

# 英文名稱皆轉大寫，方便篩選
e_name = [e.upper() for e in e_name]
group_name = pd.read_excel("D:/大學/四下/社群媒體分析/報告/K-POP藝人清單.xlsx", sheet_name = "groups")
k_group = group_name["group (korean)"].tolist()
e_group = group_name["group (english)"].tolist()
e_group = [e.upper() for e in e_group]
```

```python
from itertools import combinations

# 包含"成員" 和 "團體"
cooperations = []

# 只包含"成員"
idol_cop = []

for group in list(data.keys()):
    for video in data[group]["shorts"]:
        
        # 影片 hashtags 出現成員或團體
        idol = []
        for hashtag in video["hashtags"]:
            
            # 成員
            # 英文名稱則轉大寫放入 idol
            if hashtag[1:].upper() in e_name:
                idol.append(hashtag[1:].upper())
                
             # 韓文名稱用 excel 找英文名稱後轉大寫放入 idol
            if hashtag[1:] in k_name:
                row_index = idol_name[idol_name["name (korean)"] == hashtag[1:]].index[0]
                e = idol_name.loc[row_index, "name (english)"].upper()
                idol.append(e)
        
        # 避免重複到同個人的名稱
        unique_idol = list(set(idol))
        
        # 合作人數兩人就不用處理
        if len(unique_idol) == 2:
            idol_cop.append(tuple(unique_idol))
            
            # 目前只採取有標記雙方合作的成員名稱的資料
            # 可能會更動
            video["idol"] = [tuple(unique_idol)]
            
        # 合作人數大於兩人則取所有兩兩組合
        if len(unique_idol) > 2:
            idol_cop += list(combinations(tuple(unique_idol), 2))
            
            # 目前只採取有標記雙方合作成員名稱的資料
            # 可能會更動
            video["idol"] = list(combinations(tuple(unique_idol), 2))
            #print(idol)
            #print(video["hashtags"])
            
        # 團體
        # 一樣的處理方式
        for hashtag in video["hashtags"]:
            if hashtag[1:].upper() in e_group:
                idol.append(hashtag[1:].upper())
            if hashtag[1:] in k_group:
                row_index = group_name[group_name["group (korean)"] == hashtag[1:]].index[0]
                e = group_name.loc[row_index, "group (english)"].upper()
                idol.append(e)
        #print(f"final idol {idol}")
        
        # 避免重複
        unique_idol = list(set(idol))
        
        # 含成員及團體
        # 未拆開成兩兩組合
        cooperations.append(tuple(unique_idol))
```

```python
# pip install pyvis
import networkx as nx
from pyvis.network import Network
from collections import Counter

G = nx.MultiGraph()
G.add_edges_from(idol_cop)

# 合作次數
# 統計每對節點間的總邊數
edge_counter = Counter()
for u, v in G.edges():
    a, b = sorted((u, v))  # 讓無向邊 (A, B) 和 (B, A) 被當作同一對
    edge_counter[(a, b)] += 1

net = Network(notebook=True)

# 讓圖動起來(靜態會很亂)
net.toggle_physics(True)
net.set_options("""
{
  "physics": {
    "enabled": true,
    "solver": "forceAtlas2Based",
    "forceAtlas2Based": {
      "gravitationalConstant": -100,
      "centralGravity": 0.01,
      "springLength": 150,
      "springConstant": 0.02,
      "damping": 0.8,
      "avoidOverlap": 1
    },
    "minVelocity": 0.75,
    "stabilization": {
      "iterations": 150,
      "updateInterval": 25
    }
  }
}
""")

# 加入節點
for node in G.nodes():
    net.add_node(
        node,
        label=node,
        font={'size': 12, 'vadjust': -40}
    )

# 加入邊（只加一次每對邊(合作組合)，並顯示合作次數）
for (u, v), count in edge_counter.items():
    net.add_edge(
        u, v,
        label=str(count),  # 顯示合作次數
        color='#222222',
        font={'size': 30},
        title=f"{u} - {v} has {count} edges",
        width=2
    )

#net.show_buttons()
net.show("cooperation.html")
```

## 效益

```python
# 每個組合(有重複)
idol = []

# 每次組合的效益
# 對應 idol
value = []

# 用來算 pr值，因此只取每個影片效益
unique_value = []

for group in list(data.keys()):
    for video in data[group]["shorts"]:
        
        video_value = {}
        video_value['views'] = video['views']
        video_value['likes'] = video['likes']
        video_value['comments'] = video['comments']
        unique_value.append(video_value)
        
        if "idol" in video:
            for idol_team in video["idol"]:
                
                # 每個組合都記錄該影片效益資料(team_value)
                # 會有重複的組合(不同影片)
                team_value = {}
                team_value['views'] = video['views']
                team_value['likes'] = video['likes']
                team_value['comments'] = video['comments']
                idol.append(idol_team) # 紀錄組合
                value.append(team_value) # 紀錄對應的影片效益
```

```python
import numpy as np

# pr 值計算
def percentile_ranks(value_list):
    value_list = np.array(value_list)
    return [np.sum(value_list < x) / len(value_list) * 100 for x in value_list]

views = []
likes = []
comments = []
for value_dict in unique_value:
    views.append(value_dict["views"])
    likes.append(value_dict["likes"])
    comments.append(value_dict["comments"])

# 每個影片效益項目的 pr 值
views_ranks = percentile_ranks(views)
likes_ranks = percentile_ranks(likes)
comments_ranks = percentile_ranks(comments)

# 最終效益
final_value = []
for index in range(0, len(unique_value)):
    f_value = 0.5*views_ranks[index]+0.3*likes_ranks[index]+0.2*likes_ranks[index]
    final_value.append(f_value)
```

```python
# 每次組合的最終效益
final_value_list = []
for v_dict in value:
    index = unique_value.index(v_dict)
    t_final_value = final_value[int(index)]
    v_dict["final_value"] = t_final_value
    final_value_list.append(t_final_value)
```

```python
from collections import defaultdict

# 存取不存在的 key 時，自動建立 [] 為值
idol_values = defaultdict(list)

# 將 value 加到對應的 idol 組合裡
for i, name in enumerate(idol):
    name = tuple(sorted(name))
    idol_values[name].append(final_value_list[i])

# 計算每個組合平均效益
idol_avg = {name: sum(vals)/len(vals) for name, vals in idol_values.items()}
```

```python
G = nx.MultiGraph()
G.add_edges_from(list(idol_avg.keys()))

net = Network(notebook=True)

# 讓圖動起來(靜態會很亂)
net.toggle_physics(True)
net.set_options("""
{
  "physics": {
    "enabled": true,
    "solver": "forceAtlas2Based",
    "forceAtlas2Based": {
      "gravitationalConstant": -100,
      "centralGravity": 0.01,
      "springLength": 150,
      "springConstant": 0.02,
      "damping": 0.8,
      "avoidOverlap": 1
    },
    "minVelocity": 0.75,
    "stabilization": {
      "iterations": 150,
      "updateInterval": 25
    }
  }
}
""")

# 加入節點
for node in G.nodes():
    net.add_node(
        node,
        label=node,
        font={'size': 12, 'vadjust': -40}
    )

# 加入邊（只加一次每對邊(合作組合)，並顯示效益）
for team, value in idol_avg.items():
    u = team[0]
    v = team[1]
    net.add_edge(
        u, v,
        label=str(round(value, 2)),  # 顯示合作次數
        color='#222222',
        font={'size': 30},
        title=f"{u} - {v} link has value {round(value, 2)}",
        width=2
    )

    
#net.show_buttons()
net.show("value.html")
```

# Link Prediction

## (jaccard + adamic_adar)/ shortest path

```python
# idol_avg 可用來取組合效益分數
# edge_counter 可用來取組合合作次數
# idol, idol_cop 是所有有重複組合
# Jaccard, adamic_adar 都用不到權重
# 權重只給最短路徑使用

# 合作次數當權重
G = nx.Graph()
for (u, v), weight in edge_counter.items():
    distance = 1 / weight
    G.add_edge(u, v, weight=distance) # 合作次數多則距離越短
    
jaccard_scores = list(nx.jaccard_coefficient(G))
adamic_adar_scores = list(nx.adamic_adar_index(G))
```

```python
# 合作可能性
pre_link_score = []
adamic_adar_dict = { (u, v): score for u, v, score in adamic_adar_scores }          
for u, v, j_score in jaccard_scores:
    a_score = adamic_adar_dict[(u, v)]
    
    # 最短路徑長
    path_length = nx.shortest_path_length(G, u, v, weight='weight')
    
    # (jaccard + adamic_adar)/ shortest path
    score = (j_score + a_score) / path_length
    pre_link_score.append(dict([((u, v), score)]))
        
pre_link_score[:5]
```

```python
from collections import defaultdict

def recommend(name, threshold = 0.1):
    cop = defaultdict(list)

    # 找出目標人物的新連結
    for p_score in pre_link_score:

        u = list(p_score.keys())[0][0]
        v = list(p_score.keys())[0][1]
        if name == u:
            cop_idol = v
            cop[cop_idol] = list(p_score.values())
        if name == v:
            cop_idol = u
            cop[cop_idol] = list(p_score.values())
        
    sorted_cop = sorted(cop.items(), key=lambda x: x[1], reverse=True)

    # 門檻可調
    # filtered_cop = [item for item in sorted_cop if float(item[1][0]) >= 1]
    top_percent = threshold
    top_n = max(1, int(len(sorted_cop) * top_percent))  # 至少取 1 個
    
    # 有通過合作可能性參數門檻的組合
    filtered_cop = sorted_cop[:top_n]

    value_cop = defaultdict(list)
    for idv in filtered_cop:
        idv_value = []
    
        # 紀錄預測效益(對方合作影片平均效益)
        for (u, v), avg in idol_avg.items():
            if u == idv[0] or v == idv[0]:
                idv_value.append(avg)
        cop_avg = sum(idv_value)/len(idv_value)
        value_cop[idv[0]] = cop_avg
    
    sorted_value_cop = sorted(value_cop.items(), key=lambda x: x[1], reverse=True)
    return sorted_value_cop[:5]
```

## jaccard, adamic_adar, shortest_path 都過某比例

```python
shortest_path = []
for u, v, j_score in jaccard_scores:
    
    # 最短路徑長
    path_length = nx.shortest_path_length(G, u, v, weight='weight')
    shortest_path.append(tuple([u, v, path_length]))
```

```python
def find_score(name, scores, bool_):
    
    cop = defaultdict(list)
    
    for score in scores:

        u = score[0]
        v = score[1]
        if name == u:
            cop_idol = v
            cop[cop_idol] = score[2]
        if name == v:
            cop_idol = u
            cop[cop_idol] = score[2]
        
    sorted_cop = sorted(cop.items(), key=lambda x: x[1], reverse=bool_)
    return sorted_cop
    
def recommend_2(name, threshold = 0.1):

    sorted_jaccard = find_score(name, jaccard_scores, True)
    sorted_adamic_adar = find_score(name, adamic_adar_scores, True)
    sorted_shortest_path = find_score(name, shortest_path, False)
    
    # 門檻可調
    # filtered_cop = [item for item in sorted_cop if float(item[1][0]) >= 1]
    top_percent = threshold
    top_n = max(1, int(len(sorted_jaccard) * top_percent))  # 至少取 1 個
    
    # 有通過合作可能性參數門檻的組合
    filtered_jaccard = sorted_jaccard[:top_n]
    filtered_adamic_adar = sorted_adamic_adar[:top_n]
    filtered_shortest_path = sorted_shortest_path[:top_n]
    
    set_jaccard = set(name for name, _ in filtered_jaccard)
    set_adamic_adar = set(name for name, _ in filtered_adamic_adar)
    set_shortest_path = set(name for name, _ in filtered_shortest_path)

    # 找出三個 list 都有的合作對象
    common = set_jaccard & set_adamic_adar & set_shortest_path
    filtered_cop = list(common)
    
    value_cop = defaultdict(list)
    
    for idv in filtered_cop:
        idv_value = []
    
        # 紀錄預測效益(對方合作影片平均效益)
        for (u, v), avg in idol_avg.items():
            if u == idv or v == idv:
                idv_value.append(avg)
        cop_avg = sum(idv_value)/len(idv_value)
        value_cop[idv] = cop_avg
    
    sorted_value_cop = sorted(value_cop.items(), key=lambda x: x[1], reverse=True)
    return sorted_value_cop[:5]
```
